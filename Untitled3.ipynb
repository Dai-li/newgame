```python
import sklearn.datasets as datasets
import pandas as pd
iris=datasets.load_iris()
df=pd.DataFrame(iris.data, columns=iris.feature_names)
y=iris.target
```

```python
from sklearn.tree import DecisionTreeClassifier
dtree=DecisionTreeClassifier()
dtree.fit(df,y)
```

```python
from sklearn.externals.six import StringIO
from IPython.display import Image
from sklearn.tree import export_graphviz
import pydotplus
dot_data = StringIO()
export_graphviz(dtree, out_file=dot_data,
               filled=True,rounded=True,
               special_characters=True)
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())
Image(graph.create_png())
```

* 第一步，在本主机上建立git repo，不去链接远端

>mkdir newgame

>cd newgame

>git init  * //会生成一个'.git'的文件（包含了配置一些东西什么的）


 做一些改动，添加文件，删除文件，在'.git'的同级目录（这些改动都会被记录下来）


// 把现有的文件夹中的改动都提交到暂存区

>git add ./



//提交到代码库（就是本地的那个文件）

>git commit -m 'first_commit'   //(第一次改动)


* 第二步，在本机上面建立连接后，在远端创建一个repo（即Repository）链接本机和远端(名字和第一步创建的那个名字一样)

在GitHub页面上，建立新的repo，复制新的repo地址，回到git页面，用git remote add origin https://-----（地址）


//查看现有的状态

>git status   

>git remote add origin https://github.com/Dai-li/newgame.git

//推送到remote远端

>git push -u origin master

//第一次用git时，需要邮箱 email和用户名 name

>git config --global user.email 你的邮箱地址

>git config --global user.name 你的用户名

*  第二步也可以这么做，直接在远端建立repo，把地址复制下来，在本机直接clone（克隆），即回到git窗口git clone https://----（地址）
